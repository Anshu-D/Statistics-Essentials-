# Parametric test 1 - T-test

#Definition of parametric test - Main assumption-data is normally distributed


```python
akhil_score = [23,21,31,20,19,35,26,22,21,19]
smith_score = [46,42,62,40,38,70,52,44,42,38]
```

```python
import numpy as np
print ("Akhil mean score:",np.mean(akhil_score))
print ("Smith mean score:",np.mean(smith_score))
```

```python
#one sample T-test
import scipy.stats
scipy.stats.ttest_1samp(akhil_score,22)
```

Greater pvalue shows that 22 is near to akhil's score

```python
#Independent sample T-test
scipy.stats.ttest_ind(akhil_score,smith_score)
```

Significantly smaller p value shows that akhil's score and smith score are not equal or closer

```python
akhil_score_second_tour = [46,42,62,40,38,70,52,44,42,38]
```

```python
#paired sample ttest
scipy.stats.ttest_rel(akhil_score,akhil_score_second_tour)
```

Here p value is less than 0.05 so we have strong evdence to reject that akhil's score and akhil's score in next tour are not same

# **Parametric test 2- Anova- Tests where the means of tw or more independent samples are significantly different.**

```python
# Assumption - Normal , same distribution and identical  distribution 
```

```python
average_score = [40,44,60,50, 48,68,55,46,44,54]
```

```python
tstat, p = scipy.stats.f_oneway(akhil_score,akhil_score_second_tour,average_score)
```

```python
print ("stat = %.3f , p = %.3f" % (tstat, p))

if p> 0.05:
  print("Same distribution of scores")
else:
  print("Different distribution of scores")
```

# **Non parametric test 1- Mann Whitney U Test- Tests whether the distributions of two or more independent samples are equal or not** 

```python
#Assumptions  - Identical distribution, observations can be ranked
```

```python
class_1_score = [91,90,81,80,76]
class_2_score = [88,86,85,84,83]
```

```python
stat, p = scipy.stats.mannwhitneyu(class_1_score,class_2_score)
print ("stat = %.3f , p = %.3f" % (stat, p))

if p> 0.05:
  print("Same distribution ")
else:
  print("Different distribution")
```

# Test of Stationarity- Time Series Analysis

```python
#Augmented Dickey-Fuller Test - null hypothesis - Series is non stationary
```

```python
#Definition of stationary time series - constant mean and variance  
```

```python
from statsmodels.tsa.stattools import adfuller
stock_price_data = [121,131,142,138,139,142,156,168,165]
stat, p , lags, obs , crit , t = adfuller(stock_price_data)
```

```python
print ("stat = %.3f , p = %.3f" % (stat, p))
if p> 0.05:
  print("Series is not Stationary ")
else:
  print("Series is not Stationary")
```
